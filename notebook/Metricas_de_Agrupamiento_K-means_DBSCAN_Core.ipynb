```python
import pandas as pd
from scipy.stats import f_oneway

# Identificar las columnas categóricas
categorical_columns = ['BALANCE_CATEGORY', 'CREDIT_LIMIT_CATEGORY']

# Lista para almacenar los resultados de ANOVA
anova_results = []

# Realizar ANOVA para cada variable categórica
for column in categorical_columns:
    # Agrupar los datos por cada valor único de la columna categórica con el argumento 'observed=False'
    groups = [group["BALANCE"].dropna() for _, group in df.groupby(column, observed=False)]
    
    # Realizar la prueba ANOVA
    anova_result = f_oneway(*groups)
    
    # Almacenar los resultados en la lista
    anova_results.append({
        "column": column,
        "F-Statistic": anova_result.statistic,
        "p-value": anova_result.pvalue
    })


# Crear un DataFrame con los resultados de ANOVA
anova_df = pd.DataFrame(anova_results)

# Filtrar las variables con p-valor < 0.05
significant_anova = anova_df[anova_df["p-value"] < 0.05]["column"].tolist()

# Mostrar las variables significativas
print("\nVariables categóricas significativas para el modelo:")
print(significant_anova)

# Mostrar todos los resultados de ANOVA
print("\nResultados de ANOVA:")
print(anova_df)
```

#### Detectar valores atipicos

```python
import pandas as pd

# Suponiendo que 'df' es tu DataFrame

# Seleccionar las columnas numéricas del dataset
numeric_columns = df.select_dtypes(include=["int64", "float64"])

# Función para detectar y reemplazar valores atípicos por la moda (o un valor de tu elección)
def handle_outliers(df, columns):
    for col in columns:
        # Calcular Q1 (primer cuartil) y Q3 (tercer cuartil)
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        
        # Calcular el rango intercuartílico (IQR)
        IQR = Q3 - Q1
        
        # Establecer los límites para detectar los valores atípicos
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Calcular la moda de la columna (sin valores atípicos)
        moda = df[col].mode()[0]  # Obtiene la moda (primer valor si hay múltiples)

        # Reemplazar los valores atípicos con la moda
        df[col] = df[col].apply(lambda x: x if lower_bound <= x <= upper_bound else moda)

    return df

# Aplicar la función a las columnas numéricas
df = handle_outliers(df, numeric_columns.columns)

# Mostrar las primeras filas del DataFrame limpio
print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

# Supongamos que 'train' es tu DataFrame
# Selección de características relevantes
relevant_columns = ['PURCHASES', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES']

# Selección de las columnas relevantes del DataFrame
train_relevant = df[relevant_columns]

# Estandarización de las variables
scaler = StandardScaler()
train_standardized = scaler.fit_transform(train_relevant)

# ===================== K-means =====================

# Paso 1: Determinación del número óptimo de clusters usando el método del codo
inertia = []
range_n_clusters = range(1, 11)  # Probamos entre 1 y 10 clusters

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(train_standardized)
    inertia.append(kmeans.inertia_)  # Almacenamos la inercia

# Graficamos el método del codo
plt.figure(figsize=(8, 6))
plt.plot(range_n_clusters, inertia, marker='o', color='b')
plt.title('Método del Codo para Determinar el Número Óptimo de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.grid(True)
plt.show()

# Número óptimo de clusters (suponiendo que el codo es en 4)
optimal_clusters = 4

# Aplicamos K-means con el número óptimo de clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
df['KMeans_Cluster'] = kmeans.fit_predict(train_standardized)

# Paso 2: Cálculo del Silhouette Score para K-means
silhouette_avg_kmeans = silhouette_score(train_standardized, df['KMeans_Cluster'])
print(f"Silhouette Score (K-means): {silhouette_avg_kmeans:.4f}")

# Visualización de los clusters K-means
plt.figure(figsize=(8, 6))
plt.scatter(train_standardized[:, 0], train_standardized[:, 1], c=df['KMeans_Cluster'], cmap='viridis', s=30)
plt.title('Visualización de los Clusters usando K-means')
plt.xlabel(relevant_columns[0])
plt.ylabel(relevant_columns[1])
plt.colorbar(label='Cluster')
plt.show()

# ===================== DBSCAN =====================

# Paso 3: Aplicamos DBSCAN con diferentes valores de epsilon y min_samples
eps_values = [0.3, 0.5, 0.7]  # Valores de epsilon
min_samples_values = [3, 5, 7]  # Valores de min_samples

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        df['DBSCAN_Cluster'] = dbscan.fit_predict(train_standardized)
        
        # Visualizamos los clusters obtenidos con DBSCAN
        plt.figure(figsize=(8, 6))
        plt.scatter(train_standardized[:, 0], train_standardized[:, 1], c=df['DBSCAN_Cluster'], cmap='viridis', s=30)
        plt.title(f'Visualización de DBSCAN (eps={eps}, min_samples={min_samples})')
        plt.xlabel(relevant_columns[0])
        plt.ylabel(relevant_columns[1])
        plt.colorbar(label='Cluster')
        plt.show()

        # Contamos cuántos outliers (-1) y clusters tenemos
        n_outliers = np.sum(df['DBSCAN_Cluster'] == -1)
        n_clusters_dbscan = len(set(df['DBSCAN_Cluster'])) - (1 if -1 in df['DBSCAN_Cluster'] else 0)
        
        print(f"DBSCAN (eps={eps}, min_samples={min_samples})")
        print(f"Número de Outliers: {n_outliers}")
        print(f"Número de Clusters: {n_clusters_dbscan}")
        print("-" * 50)
```
